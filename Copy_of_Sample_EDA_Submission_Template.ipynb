## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# -----------------------------
# DATA WRANGLING / PREPARATION
# -----------------------------

# 1. Check missing values
df.isnull().sum()

# 2. Handle missing values
df['Shipment Mode'] = df['Shipment Mode'].fillna('Unknown')
df['Dosage'] = df['Dosage'].fillna('Not Specified')
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(0)

# 3. Convert important columns to numeric
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')

# Fill any remaining missing numeric values
df['Freight Cost (USD)'] = df['Freight Cost (USD)'].fillna(0)
df['Weight (Kilograms)'] = df['Weight (Kilograms)'].fillna(0)

# 4. Convert date columns to proper format
df['Scheduled Delivery Date'] = pd.to_datetime(df['Scheduled Delivery Date'], errors='coerce')
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')

# 5. Create new useful column (Delivery Time)
df['Delivery Time (Days)'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days

# 6. Handle missing delivery time
df['Delivery Time (Days)'] = df['Delivery Time (Days)'].fillna(0)

# 7. Remove duplicate rows (if any)
df = df.drop_duplicates()

# 8. Final check
df.info()
df.head()
````

Data Wrangling Summary:

The dataset was cleaned and prepared for analysis by handling missing values, converting data types, and removing duplicates.

Freight Cost and Weight columns were converted into numeric format to ensure accurate calculations. Date columns were properly formatted to analyze delivery timelines.

A new column, Delivery Time (Days), was created to measure delivery performance.

These steps ensure that the dataset is clean, consistent, and ready for further analysis and visualization.

Data Wrangling Summary:
